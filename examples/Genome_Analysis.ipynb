```python
import sys
import os

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from kitikiplot import KitikiPlot
import pandas as pd
```

```python
df= pd.read_csv( r"C:\Users\sinch\Desktop\kitikiplot\datasets\molecular+biology+promoter+gene+sequences\promoters.data", header= None )
df.columns= ["Label", "Instance_Name", "Nucleotide_Sequence"]
df.head()
```

```python
df["Nucleotide_Sequence"]= df["Nucleotide_Sequence"].str.strip()
df.head(3)
```

```python
df= df.sample(3, random_state= 1)
df
```

```python
index= 0

plot= KitikiPlot( data= [i for i in df.iloc[index, 2]], stride= 1, window_length= 55 )

plot.plot(transpose= True, 
        cell_width= 2,
        xticks_rotation= 45, 
        cmap= {'a': '#007FFF', 't': "#fffc00", "g": "#00ff00", "c": "#960018"},
        display_legend= True,
        title= "KitikiPlot Visualization for Instance: "+df.iloc[index, 1]+", Promoter "+df.iloc[index,0])
```
