```python
import sys
import os

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from kitikiplot import KitikiPlot
import pandas as pd
```

```python
df= pd.read_csv( r"C:\Users\sinch\Desktop\kitikiplot\datasets\molecular+biology+splice+junction+gene+sequences\splice.data", header= None )
df.columns= ["Label", "Instance_Name", "Nucleotide_Sequence"]
df= df.sample(3, random_state= 1)
df["Nucleotide_Sequence"]= df["Nucleotide_Sequence"].str.strip()
df
```

```python
index= 0

plot= KitikiPlot( data= [i for i in df.iloc[index, 2]], stride= 1, window_length= len(df.iloc[index, 2]) )

plot.plot(
        figsize= (20, 0.5),
        transpose= True,
        cell_width= 2,
        cmap= {'A': '#007FFF', 'T': "#fffc00", "G": "#00ff00", "C": "#960018"},
        legend_kwargs= {"bbox_to_anchor": (1.01, 1), "loc":'upper left', "borderaxespad": 0.},
        display_legend= True,
        display_yticks= True,
        xticks_rotation= 90,
        title= "KitikiPlot Visualization for Instance: "+df.iloc[index, 1].strip()+", Label : "+df.iloc[index,0].strip())
```
