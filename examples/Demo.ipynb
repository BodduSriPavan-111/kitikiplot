```python
import os

from Bio import SeqIO

from kitikiplot import KitikiPlot

os.chdir("..")
```

```python
sequence= str(list( SeqIO.parse("datasets/Gene_Sequence/sample_ls_orchid.fasta", "fasta") )[0].seq)
sequence
```

```python
ktk= KitikiPlot( data= list(sequence), window_length= 20, stride= 1)

figure= ktk.plot(
            figsize= (20, 10),
            cell_width= 2,
            cmap= {'A': '#007FFF', 'T': "#fffc00", "G": "#00ff00", "C": "#960018"},
            transpose= True,
            return_figure= True,
            legend_kwargs= {"bbox_to_anchor": (1.01, 1), "loc":'upper left', "borderaxespad": 0.}
        )
```
