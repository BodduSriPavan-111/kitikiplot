```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from kitikiplot import KitikiPlot

os.chdir("..")
```

```python
df= pd.read_csv( r"datasets\DelhiWeather\DailyDelhiClimateTrain.csv" )
print("Shape: ", df.shape)
df.head(3)
```

```python
cols= df.columns

for each_col in cols[1:]:

    plt.figure( figsize= (20, 3))
    plt.plot(df[ each_col ])
    plt.title( each_col )
    plt.show()
```

- Considering the columns: meantemp, humidity, and windspeed for further analysis.
- Apply kitikiplot

#### meantemp

```python
l= []

for i in range( df["meantemp"].shape[0]):

    value= df["meantemp"].iloc[i]

    if value>=30:
        l.append( "Scorching" )
    
    elif value>=25:
        l.append( "Hot" )

    elif value>=20:
        l.append( "Warm" )

    elif value>=15:
        l.append( "Mild" )

    elif value>=10:
        l.append( "Chilly" )

    else:
        l.append( "Freezing" )  
```

```python
COLOR_MAP= {
    "Freezing": "lime",
    "Chilly": "aqua",
    "Mild": "cornflowerblue",
    "Warm": "gold",
    "Hot": "lightcoral",
    "Scorching": "crimson"
}
```

```python
ktk= KitikiPlot( data= l[:40] )

ktk.plot(
    figsize= (20, 10),
    cell_width= 2,
    cmap= COLOR_MAP,
    title= "",
    display_xticks= False,
    xlabel= "Sequential Sliding Windows",
    ylabel= "Date (YYYY-MM-DD)",
    yticks_values= df["date"].values.tolist()[:40],
    # display_grid= True,
    display_legend= True,
    legend_kwargs= {"title": "Temperature Category"},
    kitiki_cell_kwargs= {"linewidth": 0.3}
  )
```

```python
df["date"].values.tolist()[:100]
```

```python
ktk= KitikiPlot( data= l[:100], window_length= 50 )
ktk.plot(
    figsize= (20, 10),
    cell_width= 2,
    cmap= COLOR_MAP,
    title= " ",
    xlabel= "Sequential Sliding Windows",
    ylabel= "Temperature Category Per Frame",
    display_xticks= False,
    display_yticks= False,
    xtick_prefix= "Day",
    yticks_values= df["date"].values.tolist()[:40],
    # display_grid= True,
    display_legend= True,
    kitiki_cell_kwargs= {"linewidth": 0.1},
    legend_kwargs= {"title": "Temperature Category"}
  )
```

Observations:
- There does not exit sudden raise of temperature from Green(Freezing) to Dark Blue(Mild).
- We can observe no green(Freezing) and Yellow(Warm), and dark red(Scorching) color cells are adjacent.
- Temperature is varying only to adjacent bins.
- For example the current is Yellow(Warm) then it is either staying constant or going to Mild(Dark Blue) or Hot(Red) bin but not to Green(Freezing), Sky Blue(Chilly), Dark Red(Scorching)

#### humidity

```python
l= []

for i in range( df["humidity"].shape[0]):

    value= df["humidity"].iloc[i]

    if value>=80:
        l.append( "Scorching" )
    
    elif value>=60:
        l.append( "Hot" )

    elif value>=40:
        l.append( "Warm" )

    elif value>=20:
        l.append( "Low" )

    else:
        l.append( "Very Low" )  
```

```python
COLOR_MAP= {
    "Very Low": "aqua",
    "Low": "cornflowerblue",
    "Warm": "gold",
    "Hot": "lightcoral",
    "Scorching": "crimson"
}
```

```python
ktk= KitikiPlot( data= l[:40] )

ktk.plot(
    figsize= (20, 10),
    cell_width= 2,
    cmap= COLOR_MAP,
    title= "Humidity in Delhi",
    display_xticks= False,
    xtick_prefix= "Day",
    yticks_values= df["date"].values.tolist()[:40],
    # display_grid= True,
    display_legend= True,
    kitiki_cell_kwargs= {"linewidth": 0.3}
  )
```

```python
ktk= KitikiPlot( data= l[:100], window_length= 50 )
ktk.plot(
    figsize= (20, 10),
    cell_width= 2,
    cmap= COLOR_MAP,
    title= "Humidity in Delhi",
    display_xticks= False,
    display_yticks= False,
    xtick_prefix= "Day",
    yticks_values= df["date"].values.tolist()[:40],
    # display_grid= True,
    display_legend= True,
    kitiki_cell_kwargs= {"linewidth": 0.1}
  )
```

Observations: 
- There does not exist very rapid significant change. We cannot find blue(Low) and dark red(Scorching) adjacently placed.
- Dark red(Scorching), red(Hot) and yellow(Warm) are coming in an order.
- No sudden drop from dark red(Scorching) to yellow(Warm)
- Dark Red (Scorching) is preceeded and succeeded by Red(Hot)
