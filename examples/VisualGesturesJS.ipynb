### Initialize KitikiPlot using pd.DataFrame
Dataset contains the sequence of actions collected from hand-gesture based cursor "VisualGestures.js"<br>
URL: https://www.npmjs.com/package/@learn-hunger/visual-gestures

##### Import necessary libraries

```python
import sys
import os

# sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from kitikiplot import KitikiPlot
import pandas as pd


# os.chdir("..")
```

```python
df= pd.read_excel( "datasets/VisualGestures/VisualGestures.xlsx")
print("Shape: ", df.shape)
df.head(3)
```

```python
ktk= KitikiPlot( data= df )

ktk.plot(
         figsize= (20, 6),
         cell_width= 2,
         cmap= {"Idle": "skyblue", "No_Data": "black", "Move": "violet", "Down": "Orange", "Up": "limegreen"},
         transpose= True,
         xlabel= "Detection result per each frame",
         ylabel= "Sliding window",
         xtick_prefix= "Frame",
         ytick_prefix= "Window",
         xticks_rotation= 90,         
         title= "VisualGestures.js: Sequence of cursor based actions",
         display_legend= True 
        )  
```

Consecutive outputs of cursor down and up represents the <b>"Click"</b> operation is being performed.

```python
ktk.plot(
         figsize= (20, 6),
         cell_width= 2,
         cmap= {"Down": "Orange", "Up": "limegreen"},
         edge_color= "grey",
         fallback_color= "whitesmoke",
         transpose= True,
         xlabel= "Detection result per each frame",
         ylabel= "Sliding window",
         xtick_prefix= "Frame",
         ytick_prefix= "Window",
         xticks_rotation= 90,         
         title= "VisualGestures.js: Sequence of cursor based actions",
         display_legend= True 
        )  
```
