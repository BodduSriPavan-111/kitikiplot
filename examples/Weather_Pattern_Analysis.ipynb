```python
import sys
import os

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from kitikiplot import KitikiPlot
import pandas as pd
```

```python
df= pd.read_csv( r"C:\Users\sinch\Desktop\kitikiplot\datasets\weatherHistory\weatherHistory.csv")
print("Shape: ", df.shape)
df= df.iloc[45:65, :]
print("Shape: ", df.shape)
df.head(3)
```

```python
# df.head()
```

```python
index= 0

plot= KitikiPlot( data= df["Summary"].values.tolist(), stride= 1, window_length= 10 )

plot.plot(
        figsize= (20, 5),
        cell_width= 2,
        transpose= True,
        ytick_prefix= "Window",
        xlabel= "Time",
        ylabel= "Windows",
        xticks_values= [(str(i.hour)+" "+i.strftime("%p").lower()) for i in  pd.to_datetime(df["Formatted Date"])],
        xticks_rotation= 90, 
        cmap= "Paired",
        legend_kwargs= {"bbox_to_anchor": (1.01, 1), "loc":'upper left', "borderaxespad": 0.},
        display_legend= True,
        title= "Weather Pattern on "+str( pd.to_datetime( df.iloc[0, 0] ).date()))
```
