```python
import sys
import os

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from kitikiplot import KitikiPlot
import pandas as pd
```

```python
df= pd.read_csv( r"C:\Users\sinch\Desktop\kitikiplot\datasets\weatherHistory\weatherHistory.csv")
print("Shape: ", df.shape)
df= df.iloc[:100, :]
print("Shape: ", df.shape)
df.head(3)
```

```python
index= 0

plot= KitikiPlot( data= df["Summary"].values.tolist()[:30], stride= 5, window_length= 20 )

plot.plot(
        figsize= (20, 5),
        transpose= True,
        cell_width= 2,
        yticks_values= ['a', 'aa', 'aaa'],
        xticks_rotation= 90, 
        # cmap= {'A': '#007FFF', 'T': "#fffc00", "G": "#00ff00", "C": "#960018"},
        legend_kwargs= {"bbox_to_anchor": (1.01, 1), "loc":'upper left', "borderaxespad": 0.},
        display_legend= True,
        title= "KitikiPlot Visualization for Instance: "+df.iloc[index, 1].strip()+", Label : "+df.iloc[index,0].strip())
```
